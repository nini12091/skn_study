### 중심 경향 측도

```python
import numpy as np

height = np.array([160, 170, 180, 150, 155, 165, 175, 180, 175, 170, 173, 164, 189])
```

```python
# 평균

print(np.mean(height))
```

```python
# 중위값
print(np.sort(height))
print(np.median(height))
```

```python
# 최빈값
values, count = np.unique(height, return_counts=True)
print(values, count)

max_count = np.max(count)
print(count == max_count)
print(values[count == max_count])
```

##### 산포 측도

```python
# 분산 variance
# (값 - 평균)^2의 평균


np.square(height - np.means(height))
```

```python
# np.random.normal(loc, scale, size)
# - loc : 평균
# - scale : 표준편차
# - size : 갯수 or shape
arr = np.random.normal(10, 1, 100)
# arr = np.random.normal(10, 1, (10, 100))
print(arr)

plt.hist(arr, bins=100)
plt.show
```

```python
# 균등분포
# np.random.rand
# 0.0 이상 ~ 1.0미만 수에서 균등하게 난수 추출
# - num :. 갯수 or shape

arr = np.random.rand(1000)
arr = np.random.rand(6, 3)

plt.hist(arr, bins=100)
plt.show()
```

```python
#균등분포
# np.random.randint(start, end, size)

np.random.seed(0) # 난수 계산 시작값 seed 고정

arr = np.random.randint(1, 100, 1000)
# arr = np.random.randint(1, 100, (100, 20))
print(arr)

plt.hist(arr, bins=100)
plt.show()
```

아래와 같이 학생별 과목 성적 데이터가 주어졌을 때, 각 학생의 전체 평균 성적(수학, 영어, 과학)을 계산한 후, 이 평균값에 대한 분산과 표준편차를 구하세요.

```python
# 학생별 과목 성적 데이터
scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]
```

```python
# 전체 점수를 하나의 ndarray로 생성
all_scores = np.array([scores_math, scores_english, scores_science])
print(all_scores)

# 평균
print(np.mean(all_scores, axis=0))

# 분산
print(np.var(all_scores, axis=0))

# 표준편차
print(np.std(all_scores, axis=0))
```
